```python
import matplotlib.pyplot as plt # type: ignore
import pandas as pd # type: ignore
```

```python
df = pd.read_csv('bikessales_olap.salesorder.csv')
df
```

```python
total_grossamount = df['grossamount'].sum()
total_netamount = df['netamount'].sum()

print(f'Total Gross Amount: {total_grossamount}')
print(f'Total Net Amount: {total_netamount}')
```

```python
sales_by_product = df.groupby('productid')['grossamount'].sum()
print(sales_by_product)
```

```python
total_quantity_sold = df['quantity'].sum()
print(f'Total Quantity Sold: {total_quantity_sold}')
```

```python
import matplotlib.pyplot as plt

# Visualisasi total penjualan per produk
sales_by_product.plot(kind='bar')
plt.title('Total Penjualan per Produk')
plt.xlabel('Product ID')
plt.ylabel('Total Gross Amount')
plt.show()
```

```python
quantity_by_product = df.groupby('productid')['quantity'].sum()
quantity_by_product.plot(kind='bar', color='green')
plt.title('Jumlah Penjualan (Quantity) per Produk')
plt.xlabel('Product ID')
plt.ylabel('Total Quantity Sold')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True, color='purple')
plt.title('Distribusi Harga Produk')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
